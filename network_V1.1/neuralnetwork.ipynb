```python
def filter_image(image):
    imagen=image
    model = load_model('/content/drive/MyDrive/project/version 1.1/modelV1_0_1')
    image = cv2.resize(image, (28, 28))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    image = cv2.bitwise_not(image)
    image = image / 255.0
    fig, axs = plt.subplots(1, 2, figsize=(10, 5))
    axs[0].imshow(imagen)
    axs[0].set_title('Original number')
    axs[1].imshow(image)
    axs[1].set_title(f'How neural network see it')
    for ax in axs:
        ax.axis('off')
    fig.show()
    image = np.reshape(image, (1, 28, 28, 1))
    prediction = model.predict(image)
    return np.argmax(prediction)
```

```python
imgPaths = list(paths.list_files('/content/drive/MyDrive/project/data/'))
print("Введите номер изображения:")
k = int(input())

img = cv2.imread(imgPaths[k])
filter_image(img)
```
